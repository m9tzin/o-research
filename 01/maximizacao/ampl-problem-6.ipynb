```python
# 2. Select solver
import os, sys, subprocess
caminho_mod = subprocess.check_output([sys.executable, "-m", "amplpy.modules", "path"]).decode().strip()
os.environ["PATH"] += os.pathsep + caminho_mod
from amplpy import AMPL
ampl = AMPL()
ampl.eval("option solver cbc;")
print("AMPL configurado com solver CBC")
```

```python
# 3. AMPL model
ampl.eval(
    """
var x1 >= 0;
var x2 >= 0;

maximize z: 40*x1 + 35*x2;

subject to
  capacidade: (x1/800) + (x2/600) <= 1;
  demanda_chapas: x1 <= 550;
  demanda_barras: x2 <= 580;

solve;
display x1, x2, z;
"""
)
ampl.solve()
print(f"Lucro ótimo: {ampl.get_objective('z').value()}")
print(f"x1 = {ampl.get_variable('x1').value()}")
print(f"x2 = {ampl.get_variable('x2').value()}")
```
